```python
import requests
```

```python
lines = [i.split(',') for i in requests.get('https://raw.githubusercontent.com/dm-fedorov/python_basic/master/data/opendata.stat').content.decode('UTF-8').split('\n')]
needed_sums = [int(i[3]) for i in lines if 'пенсия' in i[0] and i[1] == 'Забайкальский край' and '2018' in i[2]]
needed_dates = [i[2].replace('2018-', '') for i in lines if 'пенсия' in i[0] and i[1] == 'Забайкальский край' and '2018' in i[2]]
print(sum(needed_sums)/len(needed_sums))
```

```python
import matplotlib.pyplot as plt
plt.plot(needed_dates, needed_sums)
plt.title('Изменение размеров пенсий в Забайкальском крае в 2018 году')
plt.xlabel('Дата (месяц-число)')
plt.ylabel('Размер')
plt.grid(True)
plt.show()
```
