```python
import requests
from bs4 import BeautifulSoup
import lxml

class MyFile:

    def __init__(self, name, usage):
        self.name = name
        self.usage = usage

    def read(self):
        if self.usage != 'read':
            return "You must change file's usage to 'read'"
        else:
            with open(self.name, 'r') as f:
                return f.read()

    def write(self, content):
        if self.usage != 'write':
            return "You must change file's usage to 'write'"
        else:
            with open(self.name, 'w') as f:
                f.write(content)
                return "It's done"

    def append(self, content):
        if self.usage != 'append':
            return "You must change file's usage to 'append'"
        else:
            with open(self.name, 'a') as f:
                f.write(content)
                return "It's done"

    def read_url(self):
        return requests.get(self.name).text.encode()

    def count_urls(self):
        return len(BeautifulSoup(requests.get(self.name).text.encode(), "lxml").find_all('a', href=True))

    def write_url(self, storage):
        text = requests.get(self.name).text.encode()
        with open(storage, 'w') as f:
            f.write(str(text))
            return "It's done"
    
```
